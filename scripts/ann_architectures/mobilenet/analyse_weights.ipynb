```python
import os
import numpy as np
import matplotlib.pyplot as plt
from snntoolbox.parsing.model_libs.keras_input_lib import load

%matplotlib inline
```

```python
path = '/mnt/2646BAF446BAC3B9/Data/snn_conversion/mobilenet/v1'
```

```python
model = load(os.path.join(path, 'mobilenet_parsed'), 
             filepath_custom_objects=os.path.join(path, 'custom_objects.json'))['model']
```

```python
parameters = model.get_weights()
weights = parameters[0::2]
biases = parameters[1::2]
```

```python
plt.boxplot([np.ravel(w) for w in weights], whis=25)
plt.xlabel("Layer index")
plt.ylabel("Weight value")
plt.savefig(os.path.join(path, 'weight_distribution'))
```

```python
plt.boxplot([np.ravel(b) for b in biases])
plt.xlabel("Layer index")
plt.ylabel("Bias value")
plt.savefig(os.path.join(path, 'bias_distribution'))
```
